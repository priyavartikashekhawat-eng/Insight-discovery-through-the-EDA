```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------------------
# STEP 1: Load the dataset
# -------------------------------------------------------
# Using the same sales dataset style from Project 2 (simulating real data)
np.random.seed(42)

data = {
    "transaction_id": range(1, 101),
    "customer_id": np.random.randint(100, 140, 100),
    "product": np.random.choice(["Laptop", "Mobile", "Headphones", "Tablet"], 100),
    "quantity": np.random.randint(1, 6, 100),
    "unit_price": np.random.choice([500, 800, 1200, 25000, 45000], 100),
    "discount_pct": np.random.choice([0, 5, 10, 15, 20], 100),
    "region": np.random.choice(["North", "South", "East", "West"], 100),
    "age": np.random.randint(18, 65, 100),
}

df = pd.DataFrame(data)
df["net_amount"] = (df["quantity"] * df["unit_price"]) * (1 - df["discount_pct"] / 100)

print("Dataset Shape:", df.shape)
print(df.head(), "\n")

# -------------------------------------------------------
# STEP 2: Basic structure & summary statistics
# -------------------------------------------------------
print("Dataset Info:")
print(df.info(), "\n")

print("Summary Statistics:")
print(df.describe(), "\n")

# -------------------------------------------------------
# STEP 3: Univariate analysis (distributions)
# -------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.histplot(df["net_amount"], bins=20, kde=True)
plt.title("Distribution of Net Order Amount")
plt.xlabel("Net Amount")
plt.savefig("net_amount_distribution.png")
plt.close()

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="product", order=df["product"].value_counts().index)
plt.title("Number of Orders by Product")
plt.savefig("orders_by_product.png")
plt.close()

# -------------------------------------------------------
# STEP 4: Bivariate analysis (relationships & trends)
# -------------------------------------------------------
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="region", y="net_amount", estimator=sum, errorbar=None)
plt.title("Total Revenue by Region")
plt.savefig("revenue_by_region.png")
plt.close()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="product", y="net_amount")
plt.title("Order Value Spread by Product (Outlier Check)")
plt.savefig("order_value_by_product_boxplot.png")
plt.close()

plt.figure(figsize=(8, 6))
corr = df[["quantity", "unit_price", "discount_pct", "age", "net_amount"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.close()

print("All charts saved as PNG files.\n")

# -------------------------------------------------------
# STEP 5: Key business insights (translated into business language)
# -------------------------------------------------------
top_region = df.groupby("region")["net_amount"].sum().idxmax()
top_product = df["product"].value_counts().idxmax()
avg_discount = df["discount_pct"].mean()
high_value_pct = (df["net_amount"] > 50000).mean() * 100

print("KEY BUSINESS INSIGHTS")
print("-" * 40)
print(f"1. Highest revenue-generating region: {top_region}")
print(f"2. Most frequently ordered product: {top_product}")
print(f"3. Average discount given across orders: {avg_discount:.1f}%")
print(f"4. Percentage of high-value orders (>50,000): {high_value_pct:.1f}%")
print(f"5. Correlation between discount% and net amount: {corr.loc['discount_pct','net_amount']:.2f} "
      f"({'weak' if abs(corr.loc['discount_pct','net_amount']) < 0.3 else 
'moderate/strong'} relationship)")
```
